```python
import lightkurve as lk
```

```python
search = lk.search_lightcurve(
    "TIC 25155310",
    mission="TESS"
)

search
```

```python
lc = search.download()
```

```python
lc.plot()
```

Looking for deepest dip

```python
import numpy as np

idx = np.argmin(lc.flux)

print("Index:", idx)
print("Time:", lc.time[idx])
print("Flux:", lc.flux[idx])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))

plt.plot(
    lc.time.value,
    lc.flux.value
)

plt.xlim(
    lc.time[idx].value - 0.5,
    lc.time[idx].value + 0.5
)

plt.xlabel("Time")
plt.ylabel("Flux")
plt.title("Deepest Dip Region")

plt.show()
```
